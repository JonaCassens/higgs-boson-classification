# Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
sys.path.append('..')
from src.data.loader import load_data, load_single_dataset
```

## Load Sample Data

```python
sample_data = load_single_dataset('et', 'ggH')
print(f"Dataset shape: {sample_data.shape}")
sample_data.head()
```

## Dataset Information

```python
sample_data.info()
```

## Missing Values

```python
missing_counts = sample_data.isnull().sum()
missing_counts[missing_counts > 0]
```

## Statistical Summary

```python
sample_data.describe()
```

## Feature Distributions

```python
numeric_cols = sample_data.select_dtypes(include=[np.number]).columns
feature_cols = [col for col in numeric_cols if col != 'wt']

fig, axes = plt.subplots(len(feature_cols), 1, figsize=(10, 3*len(feature_cols)))
for i, col in enumerate(feature_cols):
    axes[i].hist(sample_data[col].dropna(), bins=50, edgecolor='black')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequency')
    axes[i].set_title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```

## Correlation Analysis

```python
correlation_matrix = sample_data[feature_cols].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Feature Correlation Matrix')
plt.tight_layout()
plt.show()
```
